```python
import matplotlib.pyplot as plt

# Data
x = ["male", "female"]
y = [17, 16]
colors = ["#03A9F4", "#E91E63"]  # Blue for male, pink for female

# Bar chart
plt.figure(figsize=(6, 5))
plt.bar(x, y, color=colors)

# Titles and labels
plt.title("Primary Check More Than 5", fontsize=14, fontweight='bold')
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Primary Check Count", fontsize=12)

# Show plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
departments = ["Urology", "Neurology", "Physiotherapy", "Critical care", "Neonatal"]
managers = [35, 30, 27, 25, 22]

# Custom colors
colors = ["#4CAF50", "#FF9800", "#03A9F4", "#E91E63", "#9C27B0"]

# Create pie chart
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(
    managers,
    labels=departments,
    colors=colors,
    autopct='%1.1f%%',
    startangle=140,
    textprops={'fontsize': 12}
)

# Title
plt.title("Top 5 Departments by Number of Managers", fontsize=14, fontweight='bold')

# Add legend
plt.legend(wedges, departments, title="Departments", loc="center left", bbox_to_anchor=(1, 0.5))

# Make it a circle
plt.axis('equal')

# Show plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Positions
positions = ["NS Supdt", "Sister IR", "Team Leader", "Head Nurse", "Nurse"]

# Corresponding values
yes_counts = [1, 3, 4, 5, 6]    # Match order with positions
no_counts  = [0, 2, 1, 1, 10]   # 0 for NS Supdt since it wasn’t in the "no" list

# X-axis positions
x = np.arange(len(positions))
width = 0.35  # Width of the bars

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar(x - width/2, yes_counts, width, label='Yes', color="#4CAF50")
bars2 = ax.bar(x + width/2, no_counts, width, label='No', color="#F44336")

# Labels and title
ax.set_xlabel('Position')
ax.set_ylabel('Count')
ax.set_title('Primary Check registered by Position (Yes vs No)')
ax.set_xticks(x)
ax.set_xticklabels(positions)
ax.legend()

# Optional: Add value labels on top of bars
for bar in bars1 + bars2:
    height = bar.get_height()
    ax.annotate(f'{height}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),  # vertical offset
                textcoords="offset points",
                ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
costs = [7000, 6000, 5000, 5000, 4500, 4000, 3500, 3000, 2500, 1800]
procedures = [
    "Fluoroscopy - Upper GI Series",
    "MRI-Spine",
    "MRI-Brain",
    "PET-CT Scan",
    "Fluoroscopy - Barium Enema",
    "MRI-Knee",
    "CT Scan-Pelvis",
    "CT Scan-Abdomen",
    "Ultrasound-Obstetric",
    "Bone Density Scan (DEXA)"
]

# Colors (distinct and vibrant)
colors = [
    "#FF6F61", "#6B5B95", "#88B04B", "#F7CAC9", "#92A8D1",
    "#955251", "#B565A7", "#009B77", "#DD4124", "#45B8AC"
]

# Create plot
plt.figure(figsize=(12, 7))
bars = plt.barh(procedures, costs, color=colors)

# Add labels
plt.xlabel("Cost (EGP)")
plt.title("Top 10 Most Expensive Medical Procedures")

# Add data labels
for bar in bars:
    width = bar.get_width()
    plt.text(width + 100, bar.get_y() + bar.get_height()/2,
             f"{width} EGP", va='center', fontsize=9)

# Flip y-axis so highest cost appears on top
plt.gca().invert_yaxis()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
positions = [
    "Junior Resident",
    "Resident",
    "Senior Resident",
    "Staff Internist",
    "Surgical Attending Physician"
]
counts = [2, 2, 2, 2, 3]

# Bright & varied colors
colors = ["#FF6F61", "#6A5ACD", "#48D1CC", "#FFD700", "#DC143C"]

# Create horizontal bar chart
plt.figure(figsize=(10, 6))
bars = plt.barh(positions, counts, color=colors)

# Add labels on the bars
for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.1, bar.get_y() + bar.get_height()/2,
             str(width), va='center', fontsize=10)

# Customize chart
plt.title("Positions Occupied by More Than One Physician", fontsize=14, fontweight='bold')
plt.xlabel("Number of Physicians")
plt.ylabel("Position")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
